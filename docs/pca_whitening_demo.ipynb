# PCA Whitening 

In image data, pixels close to each other are often correlated. By using PCA Whitening, the dimensionality of the input can be reduced and adjacent pixels can be uncorrelated.

```python
import numpy as np
from brainlit.preprocessing import whiten
from skimage import data
import matplotlib.pyplot as plt

img = data.camera()
step_size = np.array([1,1])
window_size = np.array([25,25])

whitened_img, _ = whiten(img, window_size, step_size, centered=False)
normalized = 256*(whitened_img - np.min(whitened_img))

plt.figure()
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("No Whitening")
plt.subplot(1,2,2)
plt.imshow(normalized, cmap='gray')
plt.title("Whitening")
plt.show()
```
